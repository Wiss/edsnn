# Measuring $\eta$ and $\alpha$ effect on energy equilibrium point $\breve{A}$

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```python
def equall_dep_pot_energy_level(eta: float, alpha: float = 0.5, a_h: float = 100):
    """gives A* that is achieve when depression and 
    potentiation has the same energy level"""
    if eta == 0:
        return 0
    else:
        return max(0, min(a_h*(np.log(alpha)/eta + 1), 100))
```

```python
fig, ax = plt.subplots()
tot_samp = 200
eta = np.linspace(0.01, 100, tot_samp)
alpha = np.linspace(0.05, 1.0, tot_samp)
energy = np.zeros((len(alpha), len(eta)))
for n, a in enumerate(alpha):
    for m, e in enumerate(eta):
        energy[n, m] = equall_dep_pot_energy_level(eta=e, alpha=a)
cmesh = ax.pcolormesh(alpha, eta, energy, #shading='nearest',
              shading='gouraud',
              vmin=energy.min(), vmax=energy.max())


fig.colorbar(cmesh, orientation='vertical')
#_annotate(ax, alpha, eta, "shading='nearest'")
```

```python
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset

label_s = 14
cont_s  = 10
fig, ax = plt.subplots(figsize=[5,4])

extent = alpha.min(), alpha.max(), eta.min(), eta.max()
plt.imshow(energy, interpolation='nearest', origin='lower',
           cmap='Greens', alpha=0.5, 
           aspect="auto",
           extent=extent
          )
clb = plt.colorbar()
clb.ax.set_title(r"$\breve{A}$")

contours = plt.contour(alpha, eta, energy,
                       vmin = 0., vmax = 3.,
                       levels=[10, 80, 90, 95, 97, 99],
                       color='black',
                       #cmap='Reds', 
                       alpha = .8)
plt.clabel(contours, inline=1, fontsize=cont_s)

plt.xlabel(r"$\alpha$", fontsize = label_s)
plt.ylabel(r"$\eta$", fontsize = label_s)

plt.savefig(f"alpha_eta", dpi=144,bbox_inches='tight',
           pad_inches=0)
```
